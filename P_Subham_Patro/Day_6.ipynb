```python
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

# Generating a synthetic dataset for Transformer Health: 0 (Healthy), 1 (Warning), 2 (Critical)
X_syn, y_syn = make_classification(n_samples=1000, n_features=5, n_classes=3, n_clusters_per_class=1, random_state=20)
columns = ['Oil_Temp', 'Gas_Concentration', 'Vibration', 'Acoustic_Emission', 'Load_History']

# Task 1: Create a DataFrame 'X' using X_syn and the columns list above
X = pd.DataFrame(X_syn, columns=columns)

# Task 2: Create a Series 'y' using y_syn
y = pd.Series(y_syn)

# Task 3: Split the data into X_train, X_test, y_train, y_test using a 20% test size and random_state=20
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC

# Task 4: Initialize a StandardScaler, fit it on X_train, and transform both X_train and X_test.
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Task 5: Initialize a DecisionTreeClassifier named 'dt_clf'. Set random_state=20 and max_depth=4 to prevent overfitting.
dt_clf = DecisionTreeClassifier(max_depth=4, random_state=42)

# Task 6: Initialize a Support Vector Classifier named 'svm_clf'. Set random_state=20 and kernel='linear'.
svm_clf = SVC(kernel='linear', random_state=20)
```

```python
# Task 7: Train both models using the scaled training features and training targets.
dt_clf.fit(X_train_scaled, y_train) # Train DT
svm_clf.fit(X_train_scaled, y_train) # Train SVM

# Task 8: Generate predictions on the test set for the Decision Tree model and save to 'dt_preds'
dt_preds = dt_clf.predict(X_test_scaled)

# Task 9: Generate predictions on the test set for the SVM model and save to 'svm_preds'
svm_preds = svm_clf.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

# Task 10: Calculate and print the accuracy score for the Decision Tree model
dt_acc = accuracy_score(y_test, dt_preds)
print(f"Decision Tree Accuracy: {dt_acc}")

# Task 11: Calculate and print the accuracy score for the SVM model
svm_acc = accuracy_score(y_test, svm_preds)
print(f"SVM Accuracy: {svm_acc}")

# Task 12: Generate and print the confusion matrix for the SVM model
svm_cm = confusion_matrix(y_test, svm_preds)
print("SVM Confusion Matrix:\n", svm_cm)
```

```python
# Task 13: Initialize a NEW SVM model called 'svm_rbf'. This time, use kernel='rbf' (Radial Basis Function)
svm_rbf = SVC(kernel='rbf')

# Task 14: Train this new 'svm_rbf' model on the training data
svm_rbf.fit(X_train_scaled, y_train)

# Task 15: Predict on the test set and print the new accuracy. Compare it to the linear kernel from Task 11.
rbf_preds = svm_rbf.predict(X_test_scaled)
rbf_acc = accuracy_score(y_test, rbf_preds)
print(f"SVM (RBF Kernel) Accuracy: {rbf_acc}")
```
