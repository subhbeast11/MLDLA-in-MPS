```python
import pandas as pd
import numpy as np

# We provide a simulated raw dataset of a transformer's operational parameters
data = {
    'Oil_Temp_C': [45, 50, 85, 90, 48, 52, 88, 92, 47, 89],
    'Load_Current_A': [100, 110, 400, 420, 105, 115, 410, 430, 108, 415],
    'Vibration_Hz': [50, 51, 120, 125, 50, 52, 118, 122, 51, 121],
    'Transformer_Failure': [0, 0, 1, 1, 0, 0, 1, 1, 0, 1] # 0: Normal, 1: Failure
}

# Task 1: Convert the dictionary 'data' into a Pandas DataFrame named 'df_trans'
df_trans =pd.DataFrame(data)

# Task 2: Print the first 5 rows of the DataFrame
print(df_trans.head(5))

# Task 3: Print the total number of rows and columns using the .shape attribute
shape_info = df_trans.shape
print("Dataset Shape:", shape_info)
```

```python
# Task 4: Extract the features. Create a new DataFrame 'X' containing 'Oil_Temp_C', 'Load_Current_A', and 'Vibration_Hz'.
# Hint: df_trans[['Col1', 'Col2', ...]]
X =df_trans[['Oil_Temp_C', 'Load_Current_A','Vibration_Hz']]

# Task 5: Extract the target. Create a Series 'y' containing only the 'Transformer_Failure' column.
y = df_trans['Transformer_Failure']

# Task 6: Print the shapes of X and y to confirm successful separation.
print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
```

```python
from sklearn.model_selection import train_test_split

# Task 7: Use train_test_split to split X and y into X_train, X_test, y_train, y_test.
# Set test_size to 0.3 (30% for testing) and random_state to 42 for reproducibility.
X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                    test_size=0.3,
                                                    random_state=42)

# Task 8: Print the number of samples in the training set (length of X_train).
print(len(X_train))

# Task 9: Print the number of samples in the testing set (length of X_test).
print(len(X_test))
```

```python
from sklearn.preprocessing import StandardScaler

# Task 10: Initialize the StandardScaler object as 'scaler'
scaler =StandardScaler()

# Task 11: Fit the scaler to the training data (X_train) AND transform it in one step using .fit_transform()
X_train_scaled =scaler.fit_transform(X_train)
# Task 12: Transform the testing data (X_test) using the already fitted scaler using .transform()
# Do NOT use fit_transform on the test data!
X_test_scaled = scaler.transform(X_test)
print("Original first training sample:", X_train.iloc[0].values)
print("Scaled first training sample:", X_train_scaled[0])
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Task 13: Initialize a LogisticRegression model named 'clf_model'
clf_model = LogisticRegression()

# Task 14: Train the model using the scaled training data and the training labels. Use the .fit() method.
clf_model.fit(X_train_scaled, y_train)

# Task 15: Generate predictions on the scaled testing data using the .predict() method.
# Calculate and print the accuracy score.
predictions = clf_model.predict(X_test_scaled)
accuracy = accuracy_score(y_test, predictions)
print(f"Final Model Accuracy: {accuracy * 100}%")
```
