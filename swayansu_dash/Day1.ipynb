```python
# Task 2: Find buses with undervoltage (< 0.95 p.u.)
bus_voltages_pu = {"Bus1": 1.0, "Bus2": 0.92, "Bus3": 0.98, "Bus4": 0.91}

undervoltage_buses = []

for bus, voltage in bus_voltages_pu.items():
    if voltage < 0.95:
        undervoltage_buses.append(bus)

print("Undervoltage Buses:", undervoltage_buses)
```

```python
import numpy as np

# You are given a NumPy array representing solar power generation in kW over 5 hours.
solar_generation = np.array([0.5, 1.2, 3.4, 4.1, 2.8])

# Task 3: Calculate the total energy generated.
total_energy = np.sum(solar_generation)
print("Total Energy:", total_energy)

# Task 4: Find the maximum power output.
peak_power = np.max(solar_generation)
print("Peak Power:", peak_power)

# Task 5: The sensor had a 10% calibration error. Multiply the array by 1.1 to correct it.
corrected_generation = solar_generation * 1.1
print("Corrected Generation:", corrected_generation)
```

```python
# Task 6: Extract the load values from hours 18 to 21 (indices 17 to 20) from this 24-hour load array.
daily_load_mw = np.array([10, 11, 10, 9, 9, 12, 15, 18, 20, 22, 23, 22, 21, 20, 19, 18, 20, 25, 28, 27, 24, 20, 15, 12])

evening_peak = daily_load_mw[17:21]
print("Evening Peak Load (MW):", evening_peak)

# Task 7: Find the index (hour) of the absolute maximum load in the array.
peak_hour_index = np.argmax(daily_load_mw)
print("Peak Hour Index:", peak_hour_index)
```
